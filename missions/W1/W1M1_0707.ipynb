```python
gear_am = df.groupby(['gear', 'am']).size().reset_index(name ='count')
print(gear_am)
```

```python
crosstab_result = pd.crosstab(df['gear'], df['am'])
print(crosstab_result)
```

```python
gear_am['am'] = gear_am['am'].map({0: 'Auto', 1: 'Manual'})
gear_am['gear_am'] = gear_am.apply(lambda row: f"({row['gear']}, {row['am']})", axis=1)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=gear_am, x='gear_am', y='count')
plt.title("# of Cars by Gear x Transmission")
plt.xlabel("(Gear, Auto Transmission)")
plt.ylabel("# of Cars")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.hist(figsize=(15,10), bins=10, edgecolor='black')
plt.suptitle("Histograms of Variables", fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['cyl'], df['hp'])
plt.title("Cylinder vs Horse Power")
plt.xlabel("Cylinders")
plt.ylabel("Horse Power")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['mpg'], df['hp'])
plt.title("MPG vs Horse Power")
plt.xlabel("Miles per Gallon (mpg)")
plt.ylabel("Horse Power")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Correlation Table: ")
print(df.corr(numeric_only=True))
```

```python
# Two highly correlated combinations of variables

plt.figure(figsize=(6, 4))
plt.scatter(df['wt'], df['mpg'], color='green')
plt.title("Weight vs MPG")
plt.xlabel("Weight (1000 lbs)")
plt.ylabel("Miles per Gallon (mpg)")
plt.grid(True)
plt.tight_layout()
plt.show()

# wt() vs mpg()  
# The heavier the car, the lower the fuel efficiency.
# This is because the engine requires more power.
# From the manufacturer's point of view, it is possible to improve fuel efficiency through weight optimization.
```

```python
# Two highly correlated combinations of variables

plt.figure(figsize=(6, 4))
plt.scatter(df['disp'], df['hp'], color='red')
plt.title("Displacement vs Horsepower")
plt.xlabel("Displacement (cu.in.)")
plt.ylabel("Horse Power")
plt.grid(True)
plt.tight_layout()
plt.show()

# hp() vs disp()  
# The larger the displacement, the greater the output.
# This is because larger engines can generate greater power.
# However, high displacement results in poor fuel economy, so optimal displacement design is critical.
```
